```python
#combine
files=['../input/muniraakter/pos_1.png.json','../input/muniraakter/pos_2.png.json','../input/muniraakter/pos_3.png.json']
def merge_JsonFiles(filename):
    result = list()
    for f1 in filename:
        with open(f1, 'r') as infile:
            result.append(json.load(infile))

    with open('pos_3.json', 'w') as output_file:
        json.dump(result, output_file)
merge_JsonFiles(files)

```
